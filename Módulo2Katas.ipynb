# Se inicia con la Kata del módulo 2.

## 1. Instalar paquete para crear ambientes virtuales de python

* Se ejecuta en terminal lo siguiente:

    ```
       pip install virtualenv
    ```
    Instalación del paquete ``virtualenv`` desde terminal.

![](../katasResueltas/images/kata2/console1.png)


## 2. Crear ambiente virtual con ``virtualenv``

* Se ejecuta en terminal lo siguiente:

    ```
       python -m virtualenv envk2
    ```
    Se crea ambiente virtual ``envk2`` desde terminal.

![](../katasResueltas/images/kata2/console2.png)

## 3. Activar ambiente virtual creado 

* Se ejecuta en terminal lo siguiente:

    ```
       env\Scripts\activate
    ```
    Se activa ambiente virtual ``envk2`` con ``activate`` desde terminal.

![](../katasResueltas/images/kata2/console3.png)

## 4. Visualizar paquetes instalados en el ambiente virtual

* Se ejecuta en terminal lo siguiente:

    ```
       pip freeze
    ```
    Se valida que **no** hay paquetes instalados en ambiente virtual ``envk2`` desde terminal.

![](../katasResueltas/images/kata2/console4.png)

## 5. Instalar paquete ``dateutil`` dentro del ambiente virtual

* Se ejecuta en terminal lo siguiente:

    ```
       pip install python-dateutil
    ```
    Se instala paquete ``dateutil`` en ambiente virtual ``envk2`` desde terminal.

![](../katasResueltas/images/kata2/console5.png)

## 6. Visualizar paquetes instalados en el ambiente virtual

* Se ejecuta en terminal lo siguiente:

    ```
       pip freeze
    ```
    Se valida los paquetes instalados ambiente virtual ``envk2`` desde terminal.

![](../katasResueltas/images/kata2/console6.png)
 
## 7. Salir o desactivar el ambiente

* Se ejecuta en terminal lo siguiente:

    ```
       env\Scripts\deactivate
    ```
    Se des activa ambiente virtual ``envk2`` con ``deactivate`` desde terminal.

![](../katasResueltas/images/kata2/console7.png)
